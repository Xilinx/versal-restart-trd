```python
import ipywidgets as widgets
import time
import threading
from ipywidgets import HBox, Label, VBox, Layout, GridBox, GridspecLayout, Button, Image
from IPython.display import display, HTML, Markdown, Javascript
from IPython.display import clear_output 
import vssr_trd as trd
import webbrowser


display(HTML('<style>.title {color: white ; font-size:200%}'))
display(HTML('<style>.src_doc {color: white ; font-size:150%}'))
display(HTML('<style>.curr_sel {text-align : left ; font-size: 100% ; font-weight: bold}'))
display(HTML('<style>.font_boost {color: black ; font-size:125%}'))

target_selection = widgets.RadioButtons(
    options=['APU', 'RPU'],
    style={'description_width': 'initial'},
    description='Target Selection:',
    layout=Layout(display='flex',
                 width='auto'),
    disabled=False
)

action_selection = widgets.RadioButtons(
    options=['Subsystem Restart', 'WDT Recovery', 'System Restart', 'Do Healthy Boot Test'],
    style={'description_width': 'initial'},
    description='Action Selection:',
    layout=Layout(width='auto', display='flex'),
    disabled=False
)

apu_status = widgets.ToggleButton(
    value=False,
    description='Alive',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    layout=Layout(width='auto', height='auto'),
    tooltip='Description',
)
apu_status.add_class('font_boost')

rpu_status = widgets.ToggleButton(
    value=False,
    description='Alive',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    layout=Layout(width='auto', height='auto'),
    tooltip='Description',
)
rpu_status.add_class('font_boost')

current_design = widgets.ToggleButton(
    value=False,
    description='Design: '+trd.getDesignName(),
    disabled=True,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    layout=Layout(width='auto', height='auto'),
)
current_design.add_class('font_boost')

l = widgets.Label(value='(See details on left)')

current = widgets.Label(value='')
current.add_class('curr_sel')

see_details = Button(description='(See details on left)', 
                          layout=Layout(height='auto', width='auto'),
                         disabled = True)
see_details.style.button_color = 'white'
see_details.add_class('curr_sel')

title_button = Button(description='Versal Restart Targeted Reference Design', 
                          layout=Layout(height='auto', width='auto', font_size='500%'),
                         disabled = False)
title_button.style.button_color = '#182045'
title_button.add_class('title')

sources_link = widgets.HTML(
    value="<style>a.two{background-color: #182045; color:white; \
    font-size: 175%; display: block; text-align:center; width: 100%; height: 100%; vertical-align: bottom; display: flex;\
    justify-content: center; align-items: center;}</style> \
    <a class=two href=https://github.com/Xilinx/versal-restart-trd target=_blank>Sources</a>",
)

line = Button(description='', 
                          layout=Layout(height='5px', width='auto'),
                         disabled = True)
line.style.button_color = 'black'

line2 = Button(description='', 
                          layout=Layout(height='5px', width='auto'),
                         disabled = True)
line2.style.button_color = 'black'

line3 = Button(description='', 
                          layout=Layout(height='auto', width='1px'),
                         disabled = True)
line3.style.button_color = 'gray'

line4 = Button(description='', 
                          layout=Layout(height='5px', width='auto'),
                         disabled = True)
line4.style.button_color = 'black'

line5 = Button(description='', 
                          layout=Layout(height='5px', width='auto'),
                         disabled = True)
line5.style.button_color = 'black'

line6 = Button(description='', 
                          layout=Layout(height='5px', width='auto'),
                         disabled = True)
line6.style.button_color = 'black'

documentation_link = Button(description='Documentation', 
                          layout=Layout(height='auto', width='auto'),
                         disabled = False)
documentation_link.style.button_color = '#182045'
documentation_link.add_class('src_doc')

documentation_link = widgets.HTML(
    value="<style>a.two{background-color: #182045; color:white; \
    font-size: 175%; display: block; text-align:center; width: 100%; height: 100%; vertical-align: bottom; display: flex;\
    justify-content: center; align-items: center;}</style> \
    <a class=two href=https://xilinx.github.io/versal-restart-trd/ target=_blank>Documentation</a>",
    #layout=Layout(height='100%', width='auto', padding='auto'),
)

img2 = open('images/versal.png', 'rb').read()

wi2 = widgets.Image(
    value=img2, 
    format='png', 
    layout=Layout(width='auto', height='auto')
)



click_restart = widgets.ToggleButton(
    value=False,
    description='Perform Action',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    layout=Layout(width='auto', height='auto'),
)
click_restart.add_class('font_boost')

apu_wdt = widgets.Label(value='')

rpu_wdt = widgets.Label(value='')

import markdown
from ipywidgets import HTML

gen_info = 'data/intro.md'
md1 = 'data/apu_ss.md'
md2 = 'data/apu_wdt.md'
md3 = 'data/apu_sys.md'
md4 = 'data/apu_hb_test.md'
md5 = 'data/rpu_ss.md'
md6 = 'data/rpu_wdt.md'
md7 = 'data/rpu_sys.md'
md8  = 'data/rpu_hb_test.md'

gen_html = ''

with open(gen_info, 'r') as f:
    text = f.read()
    gen_html = markdown.markdown(text, extensions=['extra', 'codehilite'])
md_dict = {
    "APU": {
        "Subsystem Restart" : md1,
        "WDT Recovery" : md2,
        "System Restart" : md3,
        "Do Healthy Boot Test" : md4
    },
    "RPU": {
        "Subsystem Restart" : md5,
        "WDT Recovery" : md6,
        "System Restart" : md7,
        "Do Healthy Boot Test" : md8
    }
}

rows = 34
cols = 13
grid = GridspecLayout(rows, cols, height='auto', width='auto')

grid[:2, 0:9] = title_button
grid[0, 9:] = sources_link
grid[1, 9:] = documentation_link
grid[2, 0:] = line2
grid[3:14, 0:8] = HTML(gen_html)
grid[5:11, 9:] = wi2
grid[14, :8] = line4
grid[14, 9:] = line6
grid[15:33, :8] = HTML(markdown.markdown(""""""))
grid[3:33, 8] = line3
grid[15, 9:] = current_design
grid[16, 9:11] = apu_status
grid[16, 11:] = rpu_status
grid[17:19, 9:] = target_selection
grid[19:21, 9:] = action_selection
grid[22, 9:] = line
grid[23, 9:] = current
grid[24, 9:] = l
grid[25, 9:] = click_restart
grid[33, 0:] = line5

def click_restart_handler(widget):
    if widget['new'] == True:
        target = target_selection.value
        action = action_selection.value
        if target == "APU":
            if action == "Subsystem Restart":
                trd.SetControl(0,0)
            elif action == "System Restart":
                trd.SetControl(0,1)
            elif action == "WDT Recovery":
                trd.SetControl(0,4)
                target_selection.disabled = True
                action_selection.disabled = True
                click_restart.disabled = True
                grid[26, 9:] = apu_wdt
                i = 1
                while True:
                    apu_wdt.value = 'Heartbeat Stopped. '+'Waiting for the APU WDT expiry: '+str(i)
                    time.sleep(1)
                    i+=1
            elif action == "Do Healthy Boot Test":
                trd.SetControl(0,0)
            else:
                print("Action Not Supported")
        elif target == "RPU":
            if action == "Subsystem Restart":
                trd.SetControl(1,0)
            elif action == "System Restart":
                trd.SetControl(1,1)
            elif action == "WDT Recovery":
                trd.SetControl(1,4)
                target_selection.disabled = True
                action_selection.disabled = True
                click_restart.disabled = True
                grid[26, 9:] = rpu_wdt
                i = 1
                while True:
                    rpu_wdt.value = 'Heartbeat Stopped. '+'Waiting for the RPU WDT expiry: '+str(i)
                    time.sleep(1)
                    i+=1
            elif action == "Do Healthy Boot Test":
                trd.SetControl(1,2)
            else:
                print("Action Not Supported")
        widget['owner'].value = False

click_restart.observe(click_restart_handler, names='value')

def work(rpu,apu):
    i = 0
    prev_action_val = ""
    prev_target_val = ""
    html = ''
    while True:
        current.value = 'Current Selection: '+str(target_selection.value)+' - '+str(action_selection.value)
        if (target_selection.value != prev_target_val) or (action_selection.value != prev_action_val):
            with open(md_dict[target_selection.value][action_selection.value], 'r') as f:
                text = f.read()
                html = markdown.markdown(text, extensions=['extra'])
            grid[15:33, :8] = HTML(html)
            prev_action_val = action_selection.value
            prev_target_val = target_selection.value
        if trd.getSubSystemStatus(1) == 1:
            rpu.description = 'RPU Alive'
            rpu.button_style = 'success'
        else:
            rpu.description = 'RPU Dead'
            rpu.button_style = 'danger'
        if trd.getSubSystemStatus(0) == 1:
            apu.description = 'APU Alive'
            apu.button_style = 'success'
        else:
            apu.description = 'APU Dead'
            apu.button_style = 'danger'
        time.sleep(1)
        i+=1

thread = threading.Thread(target=work, args=(rpu_status,apu_status))
trd.init()
thread.start() 

grid
```
